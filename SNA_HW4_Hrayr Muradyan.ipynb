### Problem 6

```python
closeness_centrality = nx.closeness_centrality(directed_graph)
betweenness_centrality = nx.betweenness_centrality(directed_graph)
eigenvector_centrality = nx.eigenvector_centrality(directed_graph, max_iter= 500)
```

```python
max_closeness = max(closeness_centrality.values())
max_betweenness = max(betweenness_centrality.values())
max_eigenvector = max(eigenvector_centrality.values())
```

```python
highest_closeness_nodes = [node for node, centrality in closeness_centrality.items() if centrality == max_closeness]
highest_betweenness_nodes = [node for node, centrality in betweenness_centrality.items() if centrality == max_betweenness]
highest_eigenvector_nodes = [node for node, centrality in eigenvector_centrality.items() if centrality == max_eigenvector]
```

```python
print(f'Highest closeness: {highest_closeness_nodes}, with value {closeness_centrality[highest_closeness_nodes[0]]}') 
print(f'Highest betweenness: {highest_betweenness_nodes}, with value {betweenness_centrality[highest_betweenness_nodes[0]]}')
print(f'Highest eigenvector: {highest_eigenvector_nodes}, with value {eigenvector_centrality[highest_eigenvector_nodes[0]]}')
```

#### Closeness: 

Closeness centrality measures how closely connected a node is to other nodes in the graph. The node with the highest closeness value of 0.2809825342704057 has the shortest average distance to all other nodes in the graph. It is positioned in a way that it can quickly reach or be reached by other nodes in terms of graph traversal.

- A higher betweenness value indicates that the node plays a crucial role in connecting different parts of the network.

#### Betweenness:

Betweenness centrality measures the extent to which a node lies on the shortest paths between other nodes in the graph. The node with the highest betweenness value of 0.3377974497301992 has a significant influence on the flow of information or interactions within the graph.

- It controls the flow of information or influence between different groups or clusters within the network.

#### Eigenvector:

Eigenvector centrality measures the influence of a node in a network based on the centrality of its neighboring nodes. The node with the highest eigenvector value of 0.07339141387266628 has connections to other nodes that are also well-connected or have high centrality in the graph. 

- It derives its own importance from being connected to other important nodes in the network.


### Problem 7

```python
from networkx.algorithms.community import greedy_modularity_communities
```

```python
communities = greedy_modularity_communities(directed_graph)
len(communities)
```

### Problem 8

```python
largest_community = max(communities, key=len)
smallest_community = min(communities, key=len)
```

```python
print(f'Largest community: {len(largest_community)}')
print(f'Smallest community: {len(smallest_community)}')
```

The largest community has 1001 nodes, which means that they are most probably tightly connected subgroup within the overall network. It represents a group of nodes that are interconnected and share strong relationships.

The smallest community has 6 nodes, which means that they represent a smaller, more isolated subset of nodes within the overall network. It has fewer connections to the rest of the network. It may suggest that the nodes have specialized or unique characteristics compared too the network.

### Problem 9

```python
largest_communities = sorted(communities, key=len, reverse=True)[:3]
```

```python
subgraphs = []
for community in largest_communities:
    subgraph = directed_graph.subgraph(community)
    subgraphs.append(subgraph)
```

```python
import matplotlib.pyplot as plt

for i, subgraph in enumerate(subgraphs):
    plt.figure(i+1)
    nx.draw(subgraph, with_labels=False, node_size=40)
    plt.title(f"Largest Community {i+1}")
    plt.show()
```

```python
degree_centrality = nx.degree_centrality(directed_graph)
```

```python
top_nodes_degree = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:3]
top_nodes_closeness = sorted(closeness_centrality, key=closeness_centrality.get, reverse=True)[:3]
top_nodes_betweenness = sorted(betweenness_centrality, key=betweenness_centrality.get, reverse=True)[:3]
top_nodes_eigenvector = sorted(eigenvector_centrality, key=eigenvector_centrality.get, reverse=True)[:3]
```

```python
G_removed = directed_graph.copy()
G_removed.remove_nodes_from(top_nodes_degree)
G_removed.remove_nodes_from(top_nodes_closeness)
G_removed.remove_nodes_from(top_nodes_betweenness)
G_removed.remove_nodes_from(top_nodes_eigenvector)
```

```python
plt.figure()
nx.draw(directed_graph, with_labels=False, node_size=30)
plt.title("Graph with Top Nodes Removed")
plt.show()
```

```python
plt.figure()
nx.draw(G_removed, with_labels=False, node_size=30)
plt.title("Graph with Top Nodes Removed")
plt.show()
```

### Problem 10

```python
influencers = []
for community in largest_communities:
    subgraph = directed_graph.subgraph(community)
    centrality = nx.degree_centrality(subgraph)
    influencer = max(centrality, key=centrality.get)
    influencers.append(influencer)
```

```python
influencer_subgraphs = []
for influencer in influencers:
    influencer_subgraph = directed_graph.subgraph(nx.ego_graph(directed_graph, influencer, radius=1))
    influencer_subgraphs.append(influencer_subgraph)
```

```python
node_sizes = []
for subgraph in influencer_subgraphs:
    centrality = nx.degree_centrality(subgraph)
    sizes = [80 if node == max(centrality, key=centrality.get) else 30 for node in subgraph.nodes()]
    node_sizes.append(sizes)
```

```python
for i, subgraph in enumerate(influencer_subgraphs):
    plt.figure(i+1)
    node_colors = ['lightblue' if node != influencers[i] else 'red' for node in subgraph.nodes()]
    nx.draw(subgraph, with_labels=False, node_color=node_colors, node_size=node_sizes[i])
    plt.title(f"Top Community {i+1}")
    plt.show()
```

The red node is the influencer (The one with the highest degree centrality)
